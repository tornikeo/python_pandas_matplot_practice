```python
def isMonotonic(array):

  #start with edgecases
  if len(array) <= 2:   #if array of size 1 or 2 then we know it's monotonic
    return True

  #then we set the direction to be the direction of the first 2 numbers

  direction = array[1] - array[0]
  for i in range(2, len(array)):  #we want to start at the 3rd number (we compared the first two numbers to determine the direction)
      if direction == 0:   #the direction could be flat (unmeaningful) for example if our array is [1, 1, 1, 3]
          direction = array[i] - array[i-1]
          continue

          if breaksDirection(direction, array[i-1], array[i]):
            return False
  return True

def breaksDirection(direction, previousInt, currentInt):

  difference = currentInt - previousInt #if our difference is equal to 0 then that just means that previousInt and currentInt are equal to each other. That doesn't break the direction
  if direction > 0:  #if our difference is positive, then our difference respects our direction...we're still trending up
      return difference < 0  #if direction is greater than 0 and difference is <0...then we're breaking our direction
  return difference > 0  #this line breaks the direction if the direction is positive
```

```python
print(isMonotonic([2, 4, 10, 39, 42, 49]))
```

```python
#Python challenges - Breadth-first Search (more difficult than depth first search)
#Out parent node will have children nodes.
#some of these children nodes will have other children nodes
#GOAL - traverse the tree level by level
#traverse each level entirely before moving on  - all all the nodes (letters) to an array

###Depth first search is different - in depth first search you traverse branch by branch


#Breadth first - go to child node then through all it's siblings
#for this we will use a Queue - first-in first out
```

```python
#### ####
class Node:

  def _init_(self, name):
    self.children = []
    self.name = name

  def addChild(self,name):
    self.children.append(Node(name))

  def breadthFirstSearch(self,array):
    queue = [self]   #initiaze it on the first node that we're calling breadth first search on
    while len(queue) > 0:   ##while the queue is not empty
      current = queue.pop(0)
      array.append(current.name)
      for child in current.children:
        queue.append(child)
    return array
```
